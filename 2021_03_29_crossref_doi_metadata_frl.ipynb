```python
import json
import requests
import urllib
from urllib.request import urlopen
```

```python
DOI = '10.21037/jlpm.2019.12.01'

resolver_cr = 'https://api.crossref.org/works/'
URL_cr = resolver_cr + DOI

resolver_pmc = 'https://www.ncbi.nlm.nih.gov/pmc/utils/idconv/v1.0/?ids='
appendix_pmc = '&idtype=doi&format=json'
URL_pmc = resolver_pmc + DOI + appendix_pmc
```

```python
url_response_cr = requests.get(URL_cr).json()['message'] 

try: 
    url_response_pmc = urllib.request.urlopen(URL_pmc).read()
    url_response_parsed_pmc = json.loads(url_response_pmc)
except:
    None
```

```python
print('type:')
print(url_response_cr['type'])
print()
print()

print('title:')
print(url_response_cr['title'])
print()
print()

print('author/s:')
try:
    for item in url_response_cr['author']:
        print(item['ORCID'])
        print(item['family'] + ", " + item['given'])
        print('-----')
        print('-----')
except:
    try:
        for item in url_response_cr['author']:
            print(item['family'] + ", " + item['given'])
            print('-----')
    except:
        None
print()
print()

print('journal:')
print(url_response_cr['issn-type'])
print(url_response_cr['container-title'])
print(url_response_cr['publisher'])
print()
print()

print('volume, issue, page:')
try:
    print(url_response_cr['volume'] + "(" + url_response_cr['issue'] + "):" + url_response_cr['page'])
except:
    try:
        print(url_response_cr['volume'] + ":" + url_response_cr['page'])
    except:
        try:
            print(url_response_cr['volume'] + "(" + url_response_cr['issue'] + ")")
        except:
            try:
                print(url_response_cr['volume'])
            except:
                None
            try:
                print(url_response_cr['issue'])
            except:
                None
            try:
                print(url_response_cr['page'])
            except:
                None
print()
print()

try:
    print('published-online:')
    print(url_response_cr['published-online'])
    print()
    print()
except:
    None


try:
    print('license:')
    for item in url_response_cr['license']:
        print(item['URL'])
    print()
    print()
except:
    None

try:
    print('language:')
    print(url_response_cr['language'])
    print()
    print()
except:
    None

try:
    print('subject:')
    print(url_response_cr['subject'])
    print()
    print()
except:
    None

print('DOI:')
print('https://doi.org/' + url_response_cr['DOI'])
print(url_response_cr['URL'])
print()
print()

try:
    print('PMC:')
    for item in url_response_parsed_pmc['records']:
        print('https://www.ncbi.nlm.nih.gov/pmc/articles/' + item['pmcid'] + '/')
    print()
    print()
except:
    None

try:
    print('funder:')
    try:
        for item in url_response_cr['funder']:
            print('https://doi.org/' + item['DOI'])
            print(item['name'])
            print(item['award'])
            print('-----')
            print('-----')
    except:
        try:
            for item in url_response_cr['funder']:
                print('https://doi.org/' + item['DOI'])
                print(item['name'])
                print('-----')
                print('-----')
        except:
            try:
                for item in url_response_cr['funder']:
                    print(item['name'])
                    print('-----')
            except:
                None
except:
    None
```
